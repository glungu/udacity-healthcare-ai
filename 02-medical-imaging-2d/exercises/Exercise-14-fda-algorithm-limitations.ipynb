```python
import pandas as pd
import numpy as np
import sklearn.metrics as m
```

Read in labels and performance data:

```python
data = pd.read_csv('labels_and_performance.csv')
data
```

First, look at the overall performance of the algorithm for the detection of pneumonia using confusion matrix. What is the specificity and sensitivity?

```python
## YOUR CODE HERE

cm = m.confusion_matrix(data['Pneumonia'], data['algorithm_output'])
print(cm)
print(f'Sensitivity (recall): {cm[0][0]/(cm[0][0] + cm[1][0])}')
print(f'Specificity: {cm[1][1]/(cm[0][1] + cm[1][1])}')
```

Now, look at the algorithm's performance in the **presence of the other diseases**. What is the specificity and sensitivity? 

```python
## YOUR CODE HERE

data2 = data[data[['Atelectasis','Effusion','Pneumothorax','Infiltration','Cardiomegaly','Mass','Nodule']].sum(axis=1) > 0]
data2
```

```python
cm = m.confusion_matrix(data2['Pneumonia'], data2['algorithm_output'])
print(cm)
print(f'Sensitivity (recall): {cm[0][0]/(cm[0][0] + cm[1][0])}')
print(f'Specificity: {cm[1][1]/(cm[0][1] + cm[1][1])}')
```

### Statement on algorithmic limitations:

Sensitivity worse by 3% in the presence of other diseases.
